## `Function building model and visualize answer`

```python
import numpy as np
class LR:
    def __init__(self, lr = 0.001, n_iters = 100):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
    
    def predict(self, X):
        print(X)
        # print(self.weights )
        # print( X)
        # print( self.bias)
        # y_pedict = np.dot(self.weights, X) + self.bias
        
        # return y_pedict
        return 0
        
    def fit(self, X, y):  
        n_samples, n_feature = (X.shape)[0], 1 
        self.weights = 0
        self.bias = 0 
        for epoc in range(self.n_iters):
            y_predict = np.dot(self.weights, X) + self.bias 
            dw = (1/2*n_samples) * np.dot(X.T, (y_predict - y))
            db = (1/2*n_samples) * np.sum(y_predict - y)
            self.weights -= self.lr * dw
            self.bias -= self.lr * db 
    
    
    
            
```

```python
# import important libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# to solve problem import module not found 
import import_ipynb
from Linear_Regression import LR  
```

```python
def model(X, y):
    # splite data to training and testing
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1234)
    Linear_regess = LR()
    # train model to make the equation of model  
    Linear_regess.fit(X_train, y_train)
    # test model  
    y_pred = Linear_regess.predict(5)
    # mean squar error evalute model
    MSE = np.mean((y_test - y_pred)**2) 
    return MSE, y_predict, X_train, X_test, y_train, y_test
    
    
```

```python
def show_plot(y_predict, X_train, X_test, y_train, y_test):
    fig, ax = plt.subplots(figsize=(8, 6))
    # drow training data with blue color
    ax.scatter(X_train, y_train, color='b', s=10)
    # drow red data with red color
    ax.scatter(X_test, y_test, color='r', s=10)
    # drow Prediction line green color
    ax.plot(X_test, y_predict)
    plt.show()
```

## `1- model predict price according transaction date col`

```python
X = df['transaction date']
y = df['house price of unit area']
MSE, y_predict, X_train, X_test, y_train, y_test = model(X, y)
print("mean squar error := ", MSE) 
show_plot(y_predict, X_train, X_test, y_train, y_test)
```
